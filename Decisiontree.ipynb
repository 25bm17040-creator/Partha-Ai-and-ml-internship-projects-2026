```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data = pd.read_csv("/content/Decision_Tree.csv")
data
```

```python
label_encoder = LabelEncoder()
data["Temperature"]=label_encoder.fit_transform(data["Temperature"])
data["Vibration"]=label_encoder.fit_transform(data["Vibration"])
data["Failure"]=label_encoder.fit_transform(data["Failure"])
```

```python
x=data[["Temperature","Vibration"]]
y=data["Failure"]
print(y)
```

```python
model = DecisionTreeClassifier(criterion="entropy")
model.fit(x,y)
```

```python
sample=[[1,1]]
```

```python
prediction=model.predict(sample)
print(prediction)
```

```python
if prediction  == 1:
  print("The machine will fail")
else:
  print("The machine will not fail")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
tree.plot_tree(
    model,
    feature_names=["Temperature","Vibration"],
    class_names=["High","Medium"]
)
plt.title("Decision Tree")
plt.show()
```
